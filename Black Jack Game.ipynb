```python
# Imports and Global Variables

import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
# Creating a Card Class

class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
#Creating Deck Class

class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        new_str = ''
        for card in self.deck:
            new_str += '\n' + card.__str__()
        return new_str

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop()
```

```python
# Creating Hand Class

class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank] 
        if card.rank == 'Ace':
            self.aces += 1
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
# Creating Chips Class

class Chips:
    
    def __init__(self, total=100):
        self.total = total  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
# Writing a function for taking bets with use try/except

def take_bet(chips):
    
    while True:
        
        try:
            chips.bet = int(input("Please enetr the amount of your bet: "))
            
        except:
            print("Sorry it isn't a number, provide an integer!")
            
        else:
            if chips.bet > chips.total:
                print("Sorry, you haven't got enough chips, you have {}".format(chips.total))
            else:
                break
            
```

```python
# Writing a function for taking hits

def hit(deck,hand):
    
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
    
```

```python
# Writing a function prompting the Player to Hit or Stand

def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while True:
        
        x = input("Do you want Hit or Stand, provide h or s: ")
        
        if x[0].lower() == 'h':
            hit(deck,hand)
        elif x[0].lower() == 's':
            print("Player Stands, Dealer turn!")
            # Powinno być:
            playing = False
            # A ja miałam:
            # continue
        else:
            print("Sorry I don't undestand, provide only h or s.")
            # Brakuje mi:
            continue
        break
```

```python
# Wrining functions to display cards

def show_some(player,dealer):
    
    print("\n Dealer's hand: ")
    print("First card hidden!")
    print(dealer.cards[1])
    
    print("\n Player's hand: ", *player.cards, sep='\n')
    
def show_all(player,dealer):
    
    print("\n Dealer's hand: ", *dealer.cards, sep='\n')
    print("Dealer's card calue: {}".format(dealer.value))
    
    print("\n Player's hand: ", *player.cards, sep='\n')
    print("Player's card calue: {}".format(player.value))
```

```python
# Writing functions to handle end of game scenarios

def player_busts(player, dealer, chips):
    print("Player busted!")
    chips.lose_bet()

def player_wins(player, dealer, chips):
    print("Player wins!")
    chips.win_bet()

def dealer_busts(player, dealer, chips):
    print("Dealer busted!")
    chips.win_bet()
    
def dealer_wins(player, dealer, chips):
    print("Dealer wins!")
    chips.lose_bet()
    
def push(player, dealer):
    print("Player and Dealer tie! Push")
```

### Structure of the game:

```python
while True:
    # Printing an opening statement
    print("Welcome in BlackJack!")
    
    # Creating & shuffling the deck, dealing two cards to each player
    deck = Deck()
    deck.shuffle()
    
    player_hand = Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
        
    # Setting up the Player's chips
    player_chips = Chips()
    
    # Prompting the Player for their bet
    take_bet(player_chips)
    
    # Showing cards (keeping one dealer card hidden)
    show_some(player_hand, dealer_hand)
    
    while playing:  # recalling this variable from our hit_or_stand function
        
        # Prompting for Player to Hit or Stand
        hit_or_stand(deck, player_hand)
        
        # Showing cards (but keeping one dealer card hidden)
        show_some(player_hand, dealer_hand)
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_hand.value > 21:
            player_busts(player_hand, dealer_hand, player_chips)
            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player_hand.value <= 21:
        while dealer_hand.value < 17:
            hit(deck, dealer_hand)
    
        # Showing all cards
        show_all(player_hand, dealer_hand)
        
        # Runnning different winning scenarios
        if dealer_hand.value > 21:
            dealer_busts(player_hand, dealer_hand, player_chips)
        elif player_hand.value < dealer_hand.value:
            dealer_wins(player_hand, dealer_hand, player_chips)
        elif player_hand.value > dealer_hand.value:
            player_wins(player_hand, dealer_hand, player_chips)
        else:
            push(player_hand, dealer_hand)
    
    # Informing Player of their chips total 
    print("Total value of the Player chips is {}".format(player_chips.total))
    
    # Asking to play again
    x = input("Do you want to play again, enter y/n: ")
    if x[0].lower() == 'y':
        playing = True
        continue      
    else:
        print("Thank you for playing!")
        break
```
